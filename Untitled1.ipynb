```python
import pandas as pd
import glob
%matplotlib inline
import plotly.express as px
```

```python
files=glob.glob('Blight.2019-09-06.2019-12-09/*urr*.csv')
df=pd.concat([pd.read_csv(path,sep=';') for path in files])
df=df.reset_index(drop=True)
df['Date']=pd.to_datetime(df['Date'])
df.head()
```

```python



min_dates=df.groupby('League')['Date'].min()
df=df.set_index('League')
df['Date']=df['Date']-min_dates
```

```python
import seaborn as sns
data=df.reset_index()
g = sns.FacetGrid(data, col="Get",hue='League',col_wrap=5,sharey=False)
g.map(sns.lineplot,'Date','Value')

```

```python
df[df.Get=='Divine Orb'].reset_index().set_index(['Date','League'])['Value'].unstack().plot(figsize=(15,5))
```

```python
df=pd.read_csv('raw_exa_chaos.csv')
df['diffstock']=df.sort_values('timestamp').groupby(['id']).diff()['stock']
df.dropna()[df['diffstock']<0].head().groupby(['timestamp','pay_currency','get_currency']).apply(lambda group: (group.diffstock*group.price).sum()/group.diffstock.sum())
```

```python
from pricing import construct_timeline,get_most_recent_price
timeline = construct_timeline()
timeline.head()
```

```python
currencies = [
        "chaos",
        "exa",
        'divine',
        'alch',
        'fuse',
        'alt',
        # 'regal',
        'vaal'
    ]
import plotly.express as px
_timeline= timeline.reset_index().sort_values(['pay_currency','get_currency'])
fig = px.line(_timeline, x="timestamp", y=0, facet_row="pay_currency", facet_col="get_currency",
             category_orders={"pay_currency": currencies ,"get_currency":currencies})
fig.update_yaxes(matches=None)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_layout(
    autosize=False,
    width=1000,
    height=800,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    paper_bgcolor="LightSteelBlue",
)
fig.show()
```
