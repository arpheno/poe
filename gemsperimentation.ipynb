```python
from poe.valuation.gems.adps import adps
import pandas as pd
import numpy as np
from poe.ninja import retrieve_prices

gem_names = {
    key: value
    for key, value in enumerate(["Anomalous", "Divergent", "Phantasmal", "Superior"])
}
gem_indexes = {value: key for key, value in gem_names.items()}
gem_indexes
display = pd.options.display

display.max_columns = 1000
display.max_rows = 10000
```

```python
prices = retrieve_prices(["SkillGem", "Currency"])
gems = [
    item for price in prices.values() for item in price if item["type"] == "SkillGem"
]
relevant_gems = [
    gem
    for gem in gems
    if gem["sparkline"]["data"]
    #     if 'Enlighten' in gem['name']
    if gem["variant"] in ("3", "5/20", "5", "20", "1")
]
```

```python
df = pd.DataFrame(relevant_gems)
df.head(10)
```

```python
df.query('variant=="5/20"')
```

```python
analysis = pd.DataFrame()
analysis["profit"] = (
    df.groupby("name")["chaosValue"].max() - df.groupby("name")["chaosValue"].min()
)
analysis["cost"] = df.groupby("name")["chaosValue"].min()
analysis["profit%"] = (
    df.groupby("name")["chaosValue"].max() - df.groupby("name")["chaosValue"].min()
) / df.groupby("name")["chaosValue"].min()
```

```python
analysis = analysis.sort_values("profit", ascending=False).reset_index()
```

```python
analysis["gem_type"] = analysis["name"].apply(
    lambda x: "special"
    if any(y in x for y in ("Enhance Support", "Enlighten Support", "Empower Support"))
    else "Awakened"
    if "Awakened" in x
    else "normal"
)
```

```python
asd = df.pivot(index="name", columns="variant", values="chaosValue")
asd.head()
```

```python
assign_types = (
    lambda x: "special"
    if any(y in x for y in ("Enhance Support", "Enlighten Support", "Empower Support"))
    else "Awakened"
    if "Awakened" in x
    else "normal"
)
funcs = {
    "special": lambda row: row["4c"] * (1 / 8)
    + row["2c"] * (1 / 8)
    + row["3c"] * 6 / 8
    - row["3"],
    "Awakened": lambda row: row["6c"] * (1 / 8)
    + row["4c"] * (1 / 8)
    + row.get("5c", 0) * 6 / 8
    - row.get("5", 1000),
    "normal": lambda row: row["21c"] * (1 / 8)
    + row.get("19c", 0) * (1 / 8)
    + row["20c"] * 6 / 8
    - row["20"],
}


def map_to_value(row):
    return funcs[assign_types(row.name)](row)


asd.apply(map_to_value, axis=1).sort_values(ascending=False)
```
